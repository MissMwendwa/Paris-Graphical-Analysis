# Import all the necessary modules to start working on the downloaded file

```python
import xmltodict
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## We use the following commands to open the XML file

```python
with open('map.osm', encoding="utf8") as osm_file:
    asdict = xmltodict.parse(osm_file.read())
```

```python
es=[]
```

```python
for w in asdict['osm']['way']:
    if not 'tag' in w:
        continue
    ts=w['tag']
    if not isinstance(ts, list):
        ts=[ts]
    if not 'highway' in (m['@k'] for m in ts):
        continue
    nds=w['nd']
    for i in range(len(nds)-1):
        es.append((nds[i]['@ref'], nds[i+1]['@ref']))
        es.append((nds[i+1]['@ref'], nds[i+1]['@ref']))
```

```python
len("ps")
```

```python
ns = set ([v for v, _ in es]+[v for _, v in es])
```

```python
ns
```

```python
ps=[]
```

```python
ps=[]
for n in asdict['osm']['node']:
    nid = n['@id']
    if nid in ns:
        ps.append((nid, float(n['@lat']), float(n['@lon'])))
```

```python
ps
```

# Create the tables to display the network connections

### Create the node file as shown below

```python
with open('paris_location.csv', 'w') as node_file:
    node_file.write('id,lat,lon\n')
    for id, lat, lon in ps:
        node_file.write(f'{id}, {lat}, {lon}\n')
```

### Create the edge file as shown below

```python
with open('paris_segments.csv', 'w') as edge_file:
    edge_file.write('src, dst\n')
    for src, dst in es:
        edge_file.write(f'{src}, {dst}\n')
```
